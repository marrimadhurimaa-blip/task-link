```python
# ========================================================
# Data Science Internship - February 2026
# Task 4 : Functions with Loops - 2
# Complete Google Colab Solution
# Clean | Humanised | Error-Free | Internship Ready
# ========================================================


# ========================================================
# Problem 1: Mobile Recharge Validation System
# ========================================================

def validate_recharge():
    """
    This function validates mobile recharge amount.
    Conditions:
    - Amount must be >= 50
    - Valid plans: 199, 299, 399, 599
    """
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: â‚¹"))

        if amount < 50:
            print("âŒ Minimum recharge amount is â‚¹50. Try again.")
        elif amount not in valid_plans:
            print("âŒ Invalid plan! Valid plans are:", valid_plans)
        else:
            print("âœ… Recharge successful for â‚¹", amount)
            break


print("\n--- Problem 1: Mobile Recharge Validation System ---")
validate_recharge()


# ========================================================
# Problem 2: Inventory Reorder Alert System
# ========================================================

def inventory_alert(products):
    """
    Prints Reorder Alert if stock < 15
    Else prints Stock OK
    """
    print("\nInventory Status:")

    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: âš  Reorder Alert (Stock = {stock})")
        else:
            print(f"{item}: âœ… Stock OK (Stock = {stock})")


print("\n--- Problem 2: Inventory Reorder Alert System ---")
products = {
    "Mobile Charger": 10,
    "USB Cable": 25,
    "Power Bank": 8,
    "Earphones": 20
}

inventory_alert(products)


# ========================================================
# Problem 3: Student Result Processing System
# ========================================================

def student_result(marks):
    """
    Calculates average marks
    Returns Pass if average >= 50 else Fail
    """
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)

    print("\nAverage Marks:", avg)

    if avg >= 50:
        return "âœ… Pass"
    else:
        return "âŒ Fail"


print("\n--- Problem 3: Student Result Processing System ---")
marks = list(map(int, input("Enter student marks separated by space: ").split()))

result = student_result(marks)
print("Result:", result)


# ========================================================
# Problem 4: Cab Fare Estimator with Retry Option
# ========================================================

def calculate_fare(distance, peak):
    """
    Calculates total cab fare
    Base fare = â‚¹50
    â‚¹12 per km
    25% extra during peak hours
    """
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak:
        fare += fare * 0.25

    return fare


print("\n--- Problem 4: Cab Fare Estimator ---")

while True:
    dist = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()

    peak_hour = True if peak_input == "yes" else False

    total_fare = calculate_fare(dist, peak_hour)

    print("ğŸš• Total Cab Fare: â‚¹", round(total_fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break


# ========================================================
# Problem 5: Employee Attendance Eligibility Checker
# ========================================================

def attendance_checker(attendance):
    """
    Calculates attendance percentage
    Returns Eligible if >= 75%
    """
    present = 0

    for day in attendance:
        if day.upper() == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("\nAttendance Percentage:", percentage)

    if percentage >= 75:
        return "âœ… Eligible"
    else:
        return "âŒ Not Eligible"


print("\n--- Problem 5: Employee Attendance Eligibility Checker ---")
attendance = input("Enter attendance record (P/A without spaces, e.g. PPAPPA): ")

result = attendance_checker(attendance)
print("Eligibility:", result)


# ========================================================
# Problem 6: Password Strength Checker
# ========================================================

def password_strength(password):
    """
    Checks password strength:
    - Minimum length 8
    - At least one digit
    - At least one special char (@#$)
    """
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "âœ… Strong Password"
    else:
        return "âŒ Weak Password"


print("\n--- Problem 6: Password Strength Checker ---")
pwd = input("Enter password: ")

strength = password_strength(pwd)
print("Password Strength:", strength)


print("\nğŸ‰ ALL PROBLEMS EXECUTED SUCCESSFULLY ğŸ‰")
```
