```python
# Task 5 - Real World Function Based Problems

# Problem 1: Social Media – Post Engagement Analyzer

def post_engagement_analyzer(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Problem 2: Healthcare – Medicine Stock Alert System

def medicine_stock_alert(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


# Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):
    total = 0

    for value in rainfall_data:
        total += value

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", round(average, 2))
    print("Rainfall Status:", status)


# Problem 4: Social Media – Duplicate Account Detection

def duplicate_account_detector(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Problem 5: Healthcare – Appointment Eligibility Checker

def appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Problem 6: Agriculture – Premium Crop Price Filter

def premium_crop_filter(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Problem 7: System Monitoring – Application Health Checker

def application_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


# Problem 8: Banking – Daily Transaction Limit Checker

def transaction_limit_checker(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = sum(attendance_list)

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", round(percentage, 2))
    print("Exam Eligibility:", status)


# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill_analyzer(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill: ₹", bill)
    print("Usage Status:", usage)

# Function Calls (Testing)


print("\n--- Problem 1 Output ---")
post_engagement_analyzer([300, 450, 300])

print("\n--- Problem 2 Output ---")
medicine_stock_alert(6)

print("\n--- Problem 3 Output ---")
rainfall_checker([60, 75, 80, 70, 75], 70)

print("\n--- Problem 4 Output ---")
duplicate_account_detector(["john", "alex", "john", "maria"])

print("\n--- Problem 5 Output ---")
appointment_eligibility(21)

print("\n--- Problem 6 Output ---")
premium_crop_filter([1500, 2500, 1800, 3200, 2100])

print("\n--- Problem 7 Output ---")
application_health_checker(7)

print("\n--- Problem 8 Output ---")
transaction_limit_checker(60000)

print("\n--- Problem 9 Output ---")
attendance_checker([1,1,1,1,0,1,1,1,0,1])

print("\n--- Problem 10 Output ---")
electricity_bill_analyzer(275)
```
