```python
# Problem 1: Smart Parking Lot Management System


def smart_parking(capacity, logs):
    parked = 0

    for action in logs:
        if action == "IN":
            parked += 1
        elif action == "OUT" and parked > 0:
            parked -= 1

    print("Currently Parked Vehicles:", parked)

    if parked >= capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


print("\n--- Problem 1 Output ---")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity, logs)


# Problem 2: Online Food Delivery Time Estimator


def delivery_time_estimator(distance, traffic, weather):
    time = distance * 5   # base time: 5 minutes per km

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8

    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


print("\n--- Problem 2 Output ---")
delivery_time_estimator(8, "High", "Rainy")


# Problem 3: Movie Theatre Seat Occupancy Analyzer


def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy >= 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


print("\n--- Problem 3 Output ---")
total_seats = 200
booked_seats = [1] * 150
seat_occupancy(total_seats, booked_seats)



# Problem 4: Cloud Server Load Classification System


def server_load(cpu_readings):
    avg_load = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg_load), "%")

    if avg_load < 50:
        print("Server Status: Normal")
    elif avg_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


print("\n--- Problem 4 Output ---")
cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)



# Problem 5: Smart Classroom Resource Usage Monitor


def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


print("\n--- Problem 5 Output ---")
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_monitor(resources)



# Problem 6: Online Event Registration Capacity Controller


def event_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        print("Confirmed Registrations:", confirmed)
        print("Waitlisted Users:", waitlisted)
        print("Registration Status: Closed")


print("\n--- Problem 6 Output ---")
event_registration(100, 105)
```
