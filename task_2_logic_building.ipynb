```python
# Data Science Internship â€“ February 2026
# Task 2: Logic Building Python Programs


# Problem 1: Unique Words

print("Problem 1: Unique Words in a Sentence\n")

sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Sentence:", sentence)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

print("\n" + "="*60 + "\n")


# Problem 2: Highest Salary


print("Problem 2: Highest Salary from Employee Data\n")

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Employee Data:", employees)
print("Highest Salary:", highest_employee, "-", highest_salary)

print("\n" + "="*60 + "\n")


# Problem 3: Max & Min Values


print("Problem 3: Maximum and Minimum Values\n")

numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)

print("\n" + "="*60 + "\n")


# Problem 4: Products Above 1000


print("Problem 4: Count Products Above Price Threshold\n")

prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Prices:", prices)
print("Products above 1000:", count)

print("\n" + "="*60 + "\n")

# Problem 5: Attendance Percentage

print("Problem 5: Attendance Percentage\n")

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_count / total_days) * 100

print("Attendance Record:", attendance)
print("Attendance Percentage:", attendance_percentage)

print("\n" + "="*60 + "\n")


# Problem 6: Remove Duplicate Phone Numbers

print("Problem 6: Remove Duplicate Phone Numbers\n")

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_phone_numbers = set(phone_numbers)

print("Phone Numbers:", phone_numbers)
print("Unique phone numbers:", unique_phone_numbers)

print("\n" + "="*60 + "\n")


# Problem 7: Character Frequency

print("Problem 7: Character Frequency Count\n")

text = "python"
char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print("String:", text)
print("Character Frequency:", char_frequency)

print("\n" + "="*60 + "\n")


# Problem 8: List to Tuple

print("Problem 8: Convert List to Tuple\n")

my_list = [10, 20, 30]
my_tuple = tuple(my_list)

print("List:", my_list)
print("Tuple:", my_tuple)

print("\n" + "="*60 + "\n")



# Problem 9: Check Key Exists

print("Problem 9: Check if Key Exists in Dictionary\n")

employee_data = {
    "Bhoomika": 75000,
    "Madhurima": 68000,
    "shrika": 72000
}

key_to_check = "Bhoomika"

if key_to_check in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")

print("\n" + "="*60 + "\n")


# Problem 10: Average Marks

print("Problem 10: Calculate Average Marks\n")

marks = [70, 75, 80, 65]

total_marks = sum(marks)
average_marks = total_marks / len(marks)

print("Marks:", marks)
print("Average Marks:", average_marks)

print("\n" + "="*60 + "\n")


print("All 10 Internship Logic Building Tasks Completed Successfully!")
```
