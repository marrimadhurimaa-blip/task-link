```python
# Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name

    def control(self, action):
        if action.upper() == "ON":
            print(f"{self.name} is ON")
        else:
            print(f"{self.name} is OFF")


# Problem 2: Employee ID Card System


class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Problem 3: Mobile Contact Record


class MobileContact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Problem 4: Product Price Tag Generator


class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("\nProduct:", self.name)
        print("Price: ₹", self.price, sep="")


# Problem 5: Movie Rating Display System


class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("\nMovie:", self.name)
        print(f"Rating: {self.rating} / 5")


# Problem 6: Delivery Address Manager


class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display(self):
        print("\nDelivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# MAIN PROGRAM EXECUTION

# Problem 1 Execution
light = SmartLight("Bedroom Light")
light.control("ON")

# Problem 2 Execution
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display()

# Problem 3 Execution
contact = MobileContact("Anita", "9876543210")
contact.display()

# Problem 4 Execution
product = Product("Headphones", 2499)
product.display()

# Problem 5 Execution
movie = Movie("Inception", 4.8)
movie.display()

# Problem 6 Execution
delivery = Delivery("Suman", "Hyderabad")
delivery.display()
```
